```python
# Cleaning Data
missing_df = df.isnull().sum()
missing_df
```

```python
df = df.drop_duplicates()
df.head(8)
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
```

```python
# Exploratory Data Analysis (EDA)
summary_stats = df.describe()
print(summary_stats)
```

```python
# Univariate Analysis:
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.subplot(1, 2, 2)
sns.countplot(df['Sex'])
plt.title('Gender Distribution')
plt.show()
```

```python
# Categorical Variables:
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(df['Pclass'])
plt.title('Pclass Distribution')
plt.subplot(1, 2, 2)
sns.countplot(df['Embarked'])
plt.title('Embarked Distribution')
plt.show()
```

```python

# Survival Analysis:
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()
survival_rate_by_sex = df.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values)
plt.title('Survival Rate by Pclass')
plt.subplot(1, 2, 2)
sns.barplot(x=survival_rate_by_sex.index, y=survival_rate_by_sex.values)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Visualizing Survival by Age:
plt.figure(figsize=(9, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Survival by Age')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
sns.pairplot(df, vars=['Age', 'Fare', 'FamilySize'], hue='Survived')
plt.suptitle('Pair Plot of Age, Fare, and FamilySize')
plt.show()
```

```python
plt.figure(figsize=(9, 6))
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.')
sns.countplot(df['Title'])
plt.title('Passenger Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Survival by Title:
title_survival_rate = df.groupby('Title')['Survived'].mean()
plt.figure(figsize=(12, 5))
sns.barplot(x=title_survival_rate.index, y=title_survival_rate.values)
plt.title('Survival Rate by Title')
plt.xticks(rotation=45)
plt.show()
```
